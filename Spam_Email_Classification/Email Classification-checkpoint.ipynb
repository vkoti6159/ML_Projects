```python
import pandas as pd

data = pd.read_csv("D:\\Data Sets\\spam_email_dataset.csv", encoding="windows-1252")
data
```

```python
# Feature and target
X = data['Subject'] 
y = data['Spam Label']
```

```python
#Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
#  Vectorization
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
#  Handle class imbalance
smote = SMOTE(random_state=42)
X_train_res, y_train_res = smote.fit_resample(X_train_tfidf, y_train)
```

```python
# Model training
model = LogisticRegression()
model.fit(X_train_res, y_train_res)
```

```python
#  Prediction & evaluation
y_pred = model.predict(X_test_tfidf)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
# Example prediction
example_email = ["Win a $1000 gift card now! Click here to claim."]
example_tfidf = vectorizer.transform(example_email)
example_prediction = model.predict(example_tfidf)
print("\nExample Email Prediction: ", "Spam" if example_prediction[0] == 1 else "Not Spam")
```
